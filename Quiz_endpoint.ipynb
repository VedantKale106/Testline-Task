```python
import requests
import pandas as pd

# URL of the JSON data
url = 'https://www.jsonkeeper.com/b/LLQT'

# Fetch the JSON data
response = requests.get(url)
data = response.json()

# Extract questions data
questions_data = data['quiz']['questions']

# Convert the extracted questions data into a DataFrame
df = pd.DataFrame(questions_data)
```
