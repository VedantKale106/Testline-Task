```python
import pandas as pd
import requests

# API URL
url = "https://api.jsonserve.com/XgAgFJ"

# Fetch data from API
response = requests.get(url)

# Check if request was successful
if response.status_code == 200:
    data = response.json()  # Convert response to JSON
    df1 = pd.DataFrame(data)  # Convert JSON to DataFrame
else:
    print("Failed to fetch data:", response.status_code)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'pastquiz' is your original DataFrame
# Create the new DataFrame with selected columns
new_df = pastquiz[['quiz_id', 'trophy_level', 'final_score', 'better_than']].copy()  # Use .copy() to avoid the warning

# Convert the columns to numeric values using .loc
new_df.loc[:, 'trophy_level'] = pd.to_numeric(new_df['trophy_level'], errors='coerce')
new_df.loc[:, 'final_score'] = pd.to_numeric(new_df['final_score'], errors='coerce')
new_df.loc[:, 'better_than'] = pd.to_numeric(new_df['better_than'], errors='coerce')

# Calculate the correlation matrix
correlation_matrix = new_df.corr()

# Plot the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap of Selected Columns')
plt.show()
```

```python
import requests
import pandas as pd

# URL of the API
url = "https://api.jsonserve.com/XgAgFJ"

# Make a GET request to fetch the JSON data
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    data = response.json()

    # Create a list to store the extracted information
    extracted_data = []

    # Extract specific details and append to the list
    for item in data:
        user_id = item.get("user_id")
        final_score = item.get("final_score")
        better_than = item.get("better_than")
        quiz_title = item.get("quiz", {}).get("title")

        extracted_data.append({
            "user_id": user_id,
            "final_score": final_score,
            "better_than": better_than,
            "quiz_title": quiz_title
        })

    # Convert the list of dictionaries to a DataFrame
    df = pd.DataFrame(extracted_data)

    # Display the DataFrame

else:
    print(f"Failed to retrieve data. Status code: {response.status_code}")
```
